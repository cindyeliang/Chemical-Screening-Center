# Import modules and L4000 .tsv

```python
import pandas as pd
pd.set_option('display.max_colwidth', 3000)
import matplotlib.pyplot as plt 
import seaborn

l4000 = pd.read_csv("Downloads/UCSC-L4000-custom-library-infomation.tsv",sep="\t")
```

# Search L4000 TargetMol library for key words based on bioactivity

```python
print('Enter name of biological pathway to search L4000 library for:')
x = input()
x = x.lower()

librarySearch = l4000[l4000["Bioactivity"].str.contains(x)==True]
librarySearch
```

# Plot distribution of compounds based on clinical trial stage, mol weight

```python
# clinicalDict = {}

# for index, row in librarySearch.iterrows():
#     clinicalStage = row['Clinical Stage']
#     if clinicalStage not in clinicalDict:
#         clinicalDict[clinicalStage] = 1
#     if clinicalStage in clinicalDict:
#         clinicalDict[clinicalStage] += 1 
        
# clinicalDF = pd.DataFrame.from_dict(clinicalDict, orient='index')

# #plot pie chart of clinical trial status
# clinicalDF.plot(kind='pie', subplots = True, legend = False)


librarySearch = librarySearch.sort_values(by=['MolWt'])


fig, ax = plt.subplots(1, 2, sharey=False, sharex=False, figsize=(15,8))


ax1=seaborn.barplot(librarySearch, x="MolWt", y="Name", ax=ax[0])

# ax2= seaborn.barplot(librarySearch, x="Clinical Stage", y="Name")
ax2 = seaborn.countplot(x='Name', hue='Clinical Stage', data=librarySearch, ax=ax[1])



ax1.set_ylabel('Cpds in ascending molecular wt', fontsize=20)
ax2.set_ylabel('Clinical trial status of compounds', fontsize=20)

# ax1.set_xticklabels(ax1.get_xticklabels(), fontsize=16, rotation=90)
ax2.set_xticklabels(ax2.get_xticklabels(), fontsize=10, rotation=90)
```

# Filter candidate compounds further by clinical trial stage

```python
print('Enter name of clinical trial stage to filter', x, 'compounds for:')
y = input()

clinicalSearch = librarySearch[librarySearch['Clinical Stage'].str.contains(y)==True]
clinicalSearch
```
